```python
import types 
def ap_mofo(L1,L2):
    if isinstance(L1,pd.core.frame.DataFrame):
        E1 = list(L1.iloc[:,0])
    else:
        E1 = list(L1)
    if isinstance(L2,pd.core.frame.DataFrame):
        E2 = list(L2.iloc[:,0])
    else:
        E2 = list(L2,[0])
    res = E1 + E2
    return pd.Series(res)

```
